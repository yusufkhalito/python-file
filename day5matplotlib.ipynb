```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('RealGDP20122023.csv')
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.rename({'index':'Year', 'tyear':'Period', 'Period':'Agriculture', 'Agriculture':'CropProduction', 'CropProduction':'Livestock',
       'Livestock':'Forestry', 'Forestry':'Fishing', 'Fishing':'Industry', 'Industry':'MininingQua', 'MiningAndQuarrying':'CrudeOil',
       'CrudePetroleumAndNaturalGas':'CoalMining', 'CoalMining':'MetalOres', 'MetalOres':'Minerals',
       'QuarryingAndOtherMinerals':'Manufacturing','Manufacturing':'OilRefining', 'OilRefining':'Cement', 'Cement':'FoodBev',
       'FoodBeverageAndTobacco':'Textile', 'TextileApparelAndFootwear':'WoodProduct',
       'WoodAndWoodProducts':'PaperProduct', 'PulpPaperAndPaperProducts':'PhamaProduct',
       'ChemicalAndPharmaceuticalProducts':'NonMetalPrd', 'NonMetallicProducts':'PlasticProduct',
       'PlasticAndRubberProducts':'EEE', 'ElectricalAndElectronics':'IronSteel',
       'BasicMetalIronAndSteel':'VehiclesASS', 'MotorVehiclesAndAssembly':'OtherManu',
       'OtherManufacturing':'Gasstem', 'ElectricityGasSteamAndAirCon':'Watersupply',
       'WaterSupplySewageWaste':'Construction', 'Construction':'Services', 'Services':'Trade', 'Trade':'Accommodation',
       'AccommodationAndFoodServices':'Transport', 'TransportationAndStorage':'RoadTransport',
       'RoadTransport':'Rail', 'RailTransportAndPipelines':'WaterTransport', 'WaterTransport':'AirTransport',
       'AirTransport':'TransService', 'TransportServices':'PostServices', 'PostAndCourierServices':'ICT',
       'InformationAndCommunication':'Publishing', 'Publishing':'MusicProduction',
       'MotionPicturesSoundRecordingAndMusicProduction':'Broadcasting', 'Broadcasting':'Entertainment',
       'ArtsEntertainmentAndRecreation':'FinInsurance', 'FinanceAndInsurance':'FinInstitution',
       'FinancialInstitutions':'Insurance', 'Insurance':'RealEstate', 'RealEstate':'SciTech',
       'ProfessionalScientificAndTechnicalServices':'Adminstration',
       'AdministrativeAndSupportServicesBusinessServices':'PublicAdminstration',
       'PublicAdministration':'Education', 'Education':'HealthService', 'HumanHealthAndSocialServices':'OtherService',
       'OtherServices':'GDP2010', 'GDP at 2010 Constant Basic Prices':'Nettaxs',
       'NetTaxesOnProducts':'GDP20102', 'GDP at 2010 Constant Market Prices':'NILL'},axis = 1,inplace =True)
df.head()
```

```python
df2 = df[df['Period'] =='Annual']
```

```python
plt.style.use('ggplot')
plt.plot(df2['Year'],df2['CrudeOil'], ls ='--', lw=2.5)
```

```python
plt.style.use('ggplot')
plt.plot(df2['Year'],df2['CrudeOil'], ls = '--', lw=2.5, marker = 'o',mec = 'navy',mfc='navy')
plt.xlabel('Year')
plt.ylabel('gdp')
plt.title('Trend of crude oil gdp in millions')
plt.show()
```

```python
figure = plt.figure(figsize =(10,10))
axes = figure.add_axes([0.1,0.1,0.8,0.7])
axes.plot(df2['Year'],df2['CrudeOil'], ls = '--', lw=2.5, marker = 'o',mec = 'navy',mfc='navy')
axes.set_xlabel('Year')
axes.set_ylabel('gdp')
axes.set_title('Trend of crude oil gdp in millions')
plt.show()
```

```python
figure = plt.figure(figsize =(10,10))
axes = figure.add_axes([0.1,0.1,0.8,0.7])
axes.plot(df2['Year'],df2['CrudeOil'], ls = '--', lw=2.5, marker = 'o',mec = 'navy',mfc='navy',markevery =3,label = 'CRUDEOIL')
axes.plot(df2['Year'],df2['Agriculture'], ls = '--', lw=2.5, marker = 'o',mec = 'navy',mfc='navy',color = 'magenta',markevery =3,label='AGRIC')
axes.plot(df2['Year'],df2['Education'], ls = '--', lw=2.5, marker = 'o',mec = 'navy',mfc='navy',color = 'green',markevery =3,label='EDUCATION')
axes.plot(df2['Year'],df2['ICT'], ls = '--', lw=2.5, marker = 'o',mec = 'navy',mfc='navy',color = 'blue',markevery =3,label ='ICT')
axes.plot(df2['Year'],df2['CrudeOil'], ls = '--', lw=2.5, marker = 'o',mec = 'navy',mfc='navy',color = '#a2006d',markevery =3,label ='REALESATE')
axes.set_xlabel('Year')
axes.set_ylabel('gdp')
axes.set_title('TREND OF VARIOUS SECTORS GDP IN MILLIONS')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(2,3, figsize=(20,10))
axes[0,0].plot(df2['Year'],df2['CrudeOil'], ls = '--', lw=2.5, marker = 'o',mec = 'navy',mfc='navy',markevery =3,)
axes[0,0].set_xlabel('YEAR',color='green',fontsize = 20)
axes[0,0].set_ylabel('GDP',color='green',fontsize = 20)
axes[0,0].set_title('TREND OF CRUDE OIL OVER TIME',color='navy',fontsize = 20)
for label in axes[0,0].get_yticklabels():
    label.set_color('crimson')

axes[0,1].bar(df2['Year'],df2['Agriculture'], color = 'magenta')
```

```python
df3 = df2[['Agriculture','CrudeOil', 'ICT', 'HealthService','Education','RealEstate']]
df3.loc['Total'] = (sum(df3['Agriculture']),sum(df3['CrudeOil']),sum(df3['ICT']),sum(df3['HealthService']),sum(df3['Education']),sum(df3['RealEstate']))
df3.tail()
```

```python
fig,axes = plt.subplots(figsize=(6,6), dpi = 120)
axes.pie(df3.loc['Total'], labels=df3.columns,autopct='%1.2f%%',shadow=True,explode=[0.0,0.0,0.0,0.5,0.0,0.0])
plt.show()
                
```
