```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
plt.style.use('ggplot')
sns.displot(x=df['total_bill'],kde =True)
```

```python
sns.regplot(x=df['total_bill'],y=df['tip'])
```

```python
sns.scatterplot(x=df['total_bill'],y=df['tip'])
```

```python
sns.countplot(x=df['day'],color='navy')
```

```python
plt.figure(figsize=(12, 8))
ax = sns.countplot(x='day',data=df,color='navy')

# add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width()  / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='magenta', xytext=(0, 5),
                textcoords='offset points')
    
plt.title('Day distribution')
plt.xticks(rotation=45)
plt.legend(loc=[1,0])
plt.show
```

```python
plt.figure(figsize=(12, 8))
ax = sns.countplot(x='day',hue='sex',data=df,palette='summer')

# add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width()  / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='magenta', xytext=(0, 5),
                textcoords='offset points')
    
plt.title('Day distribution')
plt.xticks(rotation=45)
plt.legend(loc=[1,0])
plt.show()
```

```python
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='sex',y='total_bill',data=df,palette='Blues', estimator='sum')

# add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width()  / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='magenta', xytext=(0, 5),
                textcoords='offset points')
    
plt.title('Day distribution')
plt.xticks(rotation=45)
plt.legend(loc=[1,0])
plt.show()
```

```python
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='sex',y='total_bill',data=df,palette='Blues', estimator='max', hue='day')

# add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width()  / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='magenta', xytext=(0, 5),
                textcoords='offset points')
    
plt.title('Day distribution')
plt.xticks(rotation=45)
plt.legend(loc=[1,0])
plt.show()
```
